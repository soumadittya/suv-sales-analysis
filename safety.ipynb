Processing Ratings

```python
# rating_columns = ['Bharat NCAP Child Safety Rating', 'Bharat NCAP Safety Rating', 'Global NCAP Child Safety Rating', 'Global NCAP Safety Rating']
safety['Bharat NCAP Child Safety Rating'] = safety['Bharat NCAP Child Safety Rating'].str.replace(' Star', '')
safety['Bharat NCAP Safety Rating'] = safety['Bharat NCAP Safety Rating'].str.replace(' Star', '')
safety['Global NCAP Child Safety Rating'] = safety['Global NCAP Child Safety Rating'].str.replace(' Star', '')
safety['Global NCAP Safety Rating'] = safety['Global NCAP Safety Rating'].str.replace(' Star', '')
```

```python
print(safety['Bharat NCAP Child Safety Rating'].unique())
print(safety['Bharat NCAP Safety Rating'].unique())
print(safety['Global NCAP Child Safety Rating'].unique())
print(safety['Global NCAP Safety Rating'].unique())
```

```python
# converting data types of ratings to 'float'
safety['Bharat NCAP Child Safety Rating'] = safety['Bharat NCAP Child Safety Rating'].astype('float')
safety['Bharat NCAP Safety Rating'] = safety['Bharat NCAP Safety Rating'].astype('float')
safety['Global NCAP Child Safety Rating'] = safety['Global NCAP Child Safety Rating'].astype('float')
safety['Global NCAP Safety Rating'] = safety['Global NCAP Safety Rating'].astype('float')
print(safety['Bharat NCAP Child Safety Rating'].dtype)
print(safety['Bharat NCAP Safety Rating'].dtype)
print(safety['Global NCAP Child Safety Rating'].dtype)
print(safety['Global NCAP Safety Rating'].dtype)
```

Merging 'EBD' and 'Electronic Brake Distribution (EBD)'

```python
safety.loc[(safety['Electronic Brakeforce Distribution (EBD)'].isna() ), 'Electronic Brakeforce Distribution (EBD)'] = safety['EBD']
```

```python
safety['Electronic Brakeforce Distribution (EBD)'].isna().sum()
```

```python
safety['Electronic Brakeforce Distribution (EBD)'].value_counts()
```

```python
# null values will be replaced by 'NO'
safety.loc[(safety['Electronic Brakeforce Distribution (EBD)'].isna() ), 'Electronic Brakeforce Distribution (EBD)'] = 'NO'
```

```python
safety['Electronic Brakeforce Distribution (EBD)'].value_counts()
```

```python
# dropping the 'EBD' column
safety = safety.drop(['EBD'],  axis =1)
```

Drop columns with very high number of Null Values

```python
safety = safety.drop(['Front Impact Beams', 'Side Impact Beams', 'Vehicle Stability Control System'], axis = 1)
```

```python
columns_no = ['Child Safety Locks', 
              'Curtain Airbag', 
              'Electronic Stability Control (ESC)', 
              'Impact Sensing Auto Door Unlock', 
              'Knee Airbags', 
              'Passenger Airbag',
              'Side Airbag',
              'Side Airbag-Rear',
             ]
for col in columns_no:
    safety.loc[(safety[col].isna()), col] = 'NO' 
```
